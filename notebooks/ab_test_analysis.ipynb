## Problem & Context

This project simulates a typical product analytics task in a mass digital product.

The product team is considering rolling out a new feature aimed at improving user engagement and conversion. The feature simplifies the onboarding flow by reducing the number of required steps for new users. Before making a launch decision, it is necessary to evaluate whether the feature has a statistically and practically meaningful impact on key business metrics.

An A/B experiment is conducted to compare user behavior between a control group (existing product experience) and a treatment group (new feature enabled). The goal of the analysis is not only to assess statistical significance, but also to interpret the results from a business perspective and provide a clear product recommendation. The analysis focuses on decision-making under uncertainty, trade-offs between statistical rigor and business impact, and realistic constraints commonly faced in product experiments.


## Product Hypothesis

If we simplify the onboarding flow by reducing the number of required steps for new users,
then the activation rate will increase compared to the control group,
because reducing cognitive load and friction during the first session lowers the probability of early abandonment.

As a secondary effect, we expect a decrease in time-to-activation for users in the treatment group.

## Metrics Definition

### Primary Metric



- **Activation Rate** — the proportion of new users who complete the first key action during their initial session.
Activation rate directly reflects the success of the onboarding experience and is closely tied to early product value realization.

### Secondary Metrics





- **Time-to-Activation** — time from first app open to completion of the first key action. This metric helps to understand whether the simplified onboarding reduces friction and accelerates user activation.

### Guardrail Metrics

- **Early Session Drop-off Rate** — percentage of users who abandon the onboarding flow before completion. Guardrail metrics ensure that improvements in activation are not achieved at the cost of increased user frustration or confusion.

## Experiment Design

### Groups

Users are randomly assigned to two groups:

- **Control Group** — users experience the existing onboarding flow.
- **Treatment Group** — users experience the simplified onboarding flow with fewer required steps.

---

### Randomization

Users are assigned to groups using random allocation at the user level.
This ensures that both groups are statistically comparable and eliminates selection bias.

---

### Target Uplift

We expect the new onboarding flow to increase the activation rate by approximately **5% relative uplift** compared to the control group.

This effect size is considered practically meaningful from a business perspective.

---

### Minimum Detectable Effect (MDE)

The minimum detectable effect (MDE) is set to **3% relative uplift** in activation rate.

This represents the smallest effect size that would justify a product decision, considering business impact and experimentation costs.

---

### Sample Size

The required sample size is determined based on:

- baseline activation rate
- target uplift (MDE)
- significance level (α = 0.05)
- statistical power (1 - β = 0.8)

A precise calculation will be performed during the analysis stage.

---

### Experiment Duration

The experiment is expected to run for **2 weeks**, ensuring:

- sufficient sample size
- coverage of weekday and weekend behavior patterns
- reduction of short-term noise

---

### Risks & Assumptions

- The onboarding change may unintentionally increase user confusion despite reducing steps.
- External factors (e.g., traffic source changes) may influence user behavior.
- Randomization is assumed to be correctly implemented without bias.
- No significant product changes occur during the experiment period.

## Decision Criteria

The feature will be considered successful if the following conditions are met:

- **Activation Rate** in the treatment group is higher than in the control group
- the observed uplift is statistically significant at **α = 0.05**
- the effect size is practically meaningful from a business perspective
- **Early Session Drop-off Rate** does not show a significant negative change

If the primary metric improves but the guardrail metric deteriorates, the rollout decision should be reconsidered.

If the results are inconclusive, the experiment may require a longer duration or additional validation.
